### <center> <b> <span style="color:green;">Exemple des fonctions de Bessel</span> </b></center>

Les fonctions de Bessel sont une famille de solutions à l'équation différentielle de Bessel avec un ordre alpha réel ou complexe :

Entre autres utilisations, ces fonctions se posent dans des problèmes de propagation d'ondes, tels que les modes de vibration d'une tête de tambour mince. Voici un exemple de tête de tambour circulaire ancrée au bord 

```python
import numpy as np
```

```python
def drumhead_height(n, k, distance, angle, t):
    
    kth_zero = special.jn_zeros(n, k)[-1]
    return np.cos(t) * np.cos(n*angle) * special.jn(n, distance*kth_zero)
theta = np.r_[0:2*np.pi:50j]
radius = np.r_[0:1:50j]
x = np.array([r * np.cos(theta) for r in radius])
y = np.array([r * np.sin(theta) for r in radius])
z = np.array([drumhead_height(1, 1, r, theta, 0.5) for r in radius])

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes(rect=(0, 0.05, 0.95, 0.95), projection='3d')
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='RdBu_r', vmin=-0.5, vmax=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xticks(np.arange(-1, 1.1, 0.5))
ax.set_yticks(np.arange(-1, 1.1, 0.5))
ax.set_zlabel('Z')
plt.show()
```

### <center> <b> <span style="color:green;">Les transformées de Fourier avec la bibliothèque fftpack</span> </b></center>

L'analyse de Fourier est une méthode pour exprimer une fonction comme une somme de composantes périodiques, et pour récupérer le signal de ces composantes. Lorsque la fonction et sa transformée de Fourier sont remplacées par des homologues discrétisés, on parle de transformée de Fourier discrète (DFT). La DFT est devenue un pilier du calcul numérique en partie à cause d'un algorithme très rapide pour le calculer, appelé la transformée de Fourier rapide (FFT), connue de Gauss (1805) et mise au jour sous sa forme actuelle par Cooley et Tukey

L'exemple trace la FFT de la somme de deux sinus.

```python
from scipy.fft import fft, fftfreq
# Number of sample points
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N, endpoint=False)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = fft(y)
xf = fftfreq(N, T)[:N//2]
import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```

##  <center> <b> <span style="color:blue;">Le package signal de scipy </span> </b></center>

De nombreux filtres linéaires ont la propriété d'invariance par décalage. Cela signifie que l'opération de filtrage est la même à différents endroits du signal et cela implique que la matrice de filtrage peut être construite à partir de la connaissance d'une seule ligne (ou colonne) de la matrice. Dans ce cas, la multiplication matricielle peut être réalisée à l'aide de transformées de Fourier.

## Convolution ou corrélation

La fonction correlate peut prendre des tableaux ND arbitraires en entrée et renvoyer la convolution ND des deux tableaux en sortie.
Lorsque N = 2 correlate / convole peut être utilisé pour construire des filtres d'image arbitraires pour effectuer des actions telles que le flou, l'amélioration et la détection des contours d'une image. 

```python
import numpy as np
from scipy import signal, misc
import matplotlib.pyplot as plt
image = misc.face(gray = True)
w = np.zeros((50, 50))
w[0][0] = 1.0
w[49][25] = 1.0
image_new = signal.fftconvolve(image, w)
plt.figure()
plt.imshow(image)
plt.gray()
plt.title('Original image')
plt.show()
```

```python
plt.figure()
plt.imshow(image_new)
plt.gray()
plt.title('Filtered image')
plt.show()
```

## Conception de filtre

Les filtres temporels discrets peuvent être classés en filtres à réponse finie (FIR) et en filtres à réponse infinie (IIR). Les filtres FIR peuvent fournir une réponse en phase linéaire, contrairement aux filtres IIR. SciPy fournit des fonctions pour concevoir les deux types de filtres.

La fonction firwin permet de concevoir des filtres, selon les arguments fournis, la fonction renvoie différents types de filtres (par exemple, passe-bas, passe-bande…).

L'exemple ci-dessous conçoit respectivement un filtre passe-bas et un filtre coupe-bande

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
b1 = signal.firwin(40, 0.5)
b2 = signal.firwin(41, [0.3, 0.8])
w1, h1 = signal.freqz(b1)
w2, h2 = signal.freqz(b2)
plt.title('Digital filter frequency response')
plt.plot(w1, 20*np.log10(np.abs(h1)), 'b')
plt.plot(w2, 20*np.log10(np.abs(h2)), 'r')
plt.ylabel('Amplitude Response (dB)')
plt.xlabel('Frequency (rad/sample)')
plt.grid()
plt.show()
```
